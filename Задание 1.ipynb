```python
Датасет The Iris состоит из данных о 3 различных видах ирисов (Setosa, Versicolour, и Virginica), 
ширине и длине их наружной и внутренней доли, хранящихся в 150x4 numpy.ndarray
```

```python
import matplotlib.pyplot as plt
import mpl_toolkits.mplot3d  # noqa: F401

from sklearn import datasets
from sklearn.decomposition import PCA

#импортируем данные
iris = datasets.load_iris()
X = iris.data[:, :2]  #берем только два признака (длина и ширина наружной доли)
y = iris.target

x_min, x_max = X[:, 0].min() - 0.5, X[:, 0].max() + 0.5
y_min, y_max = X[:, 1].min() - 0.5, X[:, 1].max() + 0.5

plt.figure(1, 
           figsize=(int(X[:, 0].max() - X[:, 0].min() + 1)*2.5, 
                    int(X[:, 1].max() - X[:, 1].min() + 1)*2.5)
          )
plt.clf()

#Строим точки тренировочных данных
plt.scatter(X[:, 0], X[:, 1], c=y, cmap=plt.cm.Set1, edgecolor="k")
plt.xlabel("Длина наружной доли")
plt.ylabel("Ширина наружной доли")

plt.xlim(x_min, x_max)
plt.ylim(y_min, y_max)
plt.xticks(())
plt.yticks(())

#Построим в трех PCA измерениях:

fig = plt.figure(2, 
                figsize=(int(X[:, 0].max() - X[:, 0].min() + 1)*3, 
                         int(X[:, 1].max() - X[:, 1].min() + 1)*3)
                )
ax = fig.add_subplot(111, projection="3d", elev=-150, azim=110)

X_reduced = PCA(n_components=3).fit_transform(iris.data)
ax.scatter(
    X_reduced[:, 0],
    X_reduced[:, 1],
    X_reduced[:, 2],
    c=y,
    cmap=plt.cm.Set1,
    edgecolor="k",
    s=40,
)

ax.set_title("Первые три PCA измерения")
ax.set_xlabel("Длина наружной доли")
ax.w_xaxis.set_ticklabels([])
ax.set_ylabel("Ширина наружной доли")
ax.w_yaxis.set_ticklabels([])
ax.set_zlabel("Длина внутренней доли")
ax.w_zaxis.set_ticklabels([])

plt.show()
```
